```python
# Removing uneccessary columns: dont need ID Column
dataset = dataset.drop('ID', axis=1)
```

```python
# Checking for null values
dataset.isnull().sum()
```

```python
# UNIVARIATE ANALYSIS: Looking at the type of data present in different columns
cat_cols=['SEX','EDUCATION','MARRIAGE','default.payment.next.month']

fig,ax=plt.subplots(1,4,figsize=(25,5))

for cols,subplots in zip(cat_cols,ax.flatten()):
    sns.countplot(x=dataset[cols],ax=subplots)
    
# SEX: Gender (1=male, 2=female)
# EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
# MARRIAGE: Marital status (1=married, 2=single, 3=others)
# Default payment (1=yes, 0=no)
```

```python
# Vizualizing the imbalance 

yes=(((dataset['default.payment.next.month']==1).sum())/len(dataset['default.payment.next.month']))*100
no=(((dataset['default.payment.next.month']==0).sum())/len(dataset['default.payment.next.month']))*100

x=[yes,no]

plt.pie(x,labels=['Yes','No'],colors=['red', 'blue'],radius=2,autopct='%1.0f%%')
plt.title('default.payment.next.month')
plt.show()
```

```python
# Counting dataset['default.payment.next.month'].value_counts(normalize=True)
```

```python
# Removing default.payment.next.month
X=dataset.drop('default.payment.next.month',axis=1)
y=dataset['default.payment.next.month']
```

```python
# Spliting the dataset in train and data columns
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
# Training data using Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
model=rfc.fit(X_train,y_train)
```

```python
# predicting the test 
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,predictions))
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, predictions)
print(cm)
accuracy_score(y_test, predictions)
```
